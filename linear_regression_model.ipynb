```python
# Create a copy of the dataset
flights_processed = df.copy()

# Handle missing values if any
flights_processed.dropna(inplace=True)

# Split features and target variable
X = flights_processed.drop("Average_Fare", axis=1)
y = flights_processed["Average_Fare"]

# Feature scaling. The resulting features will have a mean 0 and a variance 1
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Convert back to DataFrame for better handling
X_scaled_df = pd.DataFrame(X_scaled, columns=X.columns)
```

```python
# Data properties
data = np.array(y)
mean_value = np.mean(data)
print("The mean fare price is ", mean_value)
print("The standard deviation of the fare price is ", np.std(data))

from scipy.stats import norm
data = np.array(y)
mean_value = np.mean(data)
std_dev = np.std(data)

x = np.linspace(mean_value - 4*std_dev, mean_value + 4*std_dev, 500)
pdf = norm.pdf(x, loc=mean_value, scale=std_dev)

plt.figure(figsize=(10, 6))
plt.plot(x, pdf, color='blue', linewidth=2, label='Normal Distribution PDF')
plt.hist(data, bins=30, density=True, alpha=0.6, color='skyblue', label='Histogram of y')
plt.title('Normal Distribution Curve and Histogram of Target Variable (y)')
plt.xlabel('Fare/Price (y)')
plt.ylabel('Probability Density')
plt.legend()
plt.grid(True)
plt.show()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_scaled_df, y, test_size=0.2, random_state=42)

print(f"Training set size: {X_train.shape}")
print(f"Testing set size: {X_test.shape}")
```

```python
def linear_regression_normal_equation(X, y):
    """
    Implement linear regression using the normal equation:
    θ = (X^T X)^(-1) X^T y
    """
    # Add bias term (column of ones)
    X_b = np.c_[np.ones((X.shape[0], 1)), X]

    # Calculate the coefficients using the normal equation
    theta = np.linalg.inv(X_b.T.dot(X_b)).dot(X_b.T).dot(y)

    return theta

# Convert to numpy arrays for our implementation
X_train_np = X_train.values
y_train_np = y_train.values

# Train model using normal equation
theta_normal = linear_regression_normal_equation(X_train_np, y_train_np)

print("Intercept:", theta_normal[0])
print("Coefficients:", theta_normal[1:])
```

```python
def predict(X, theta):
    """Make predictions using the trained model"""
    X_b = np.c_[np.ones((X.shape[0], 1)), X]
    return X_b.dot(theta)

# Convert test set to numpy array
X_test_np = X_test.values

# Predictions with normal equation
y_pred_normal = predict(X_test_np, theta_normal)
```

```python
from sklearn.metrics import mean_squared_error

def evaluate_model(y_true, y_pred, model_name):
    """Evaluate model performance using only Mean Squared Error"""
    mse = mean_squared_error(y_true, y_pred)
    print(f"{model_name} Performance:")
    print(f"Mean Squared Error: {mse:.2f}")
    return mse

# Evaluate normal equation model
mse_normal = evaluate_model(y_test, y_pred_normal, "Normal Equation")
```

```python
best_features = []
best_score = np.inf # Initialize for minimization
features = list(X.columns)

while features:
    mse_scores = {}
    for feature in features:
        selected = best_features + [feature]
        new_X = X[selected]

        X_train, X_test, y_train, y_test = train_test_split(new_X, y, test_size = 0.2, random_state=42)

        X_train_np = X_train.values
        y_train_np = y_train.values
        theta_normal = linear_regression_normal_equation(X_train_np, y_train_np)

        y_pred_normal = predict(X_test.values, theta_normal)
        mse = evaluate_model(y_test.values, y_pred_normal, "Normal Equation")

        mse_scores[feature] = mse

    best_feature, new_best_score = min(mse_scores.items(), key=lambda item:item[1])

    if new_best_score < best_score - 10:
        best_features.append(best_feature)
        features.remove(best_feature)
        print("Used Features: ", best_features, "| MSE:", round(new_best_score, 3), "| Improvement:", round(abs(new_best_score - best_score), 3) if not np.isinf(best_score) else round(new_best_score,3))
        best_score = new_best_score
    else:
        break
```
