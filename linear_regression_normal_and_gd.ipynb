NORMAL EQUATION

```python
def linear_regression_normal_equation(X, y):
  # Implements linear regression using the Normal Equation: Î¸ = (X^T X)^(-1) X^T y
  # Outputs model parameters (theta)

  # Add bias term (column of ones)
  X_b = np.c_[np.ones((X.shape[0], 1)), X]

  # Compute theta using the normal equation
  theta = np.linalg.inv(X_b.T @ X_b) @ X_b.T @ y

  return theta

# Extract features and target
X = df[['NonStopMiles','MktMilesFlown','Market_HHI', 'Market_share', 'MktCoupons']].values
y = df["Average_Fare"].values.reshape(-1, 1)
```

```python
# Train using the normal equation
start_train = time.time()
theta_normal = linear_regression_normal_equation(X, y)
end_train = time.time()
training_time = end_train - start_train

# Print coefficients
print("Coefficients from Normal Equation:")
print(f"Intercept:     {theta_normal[0][0]:.4f}")
print(f"NonStopMiles:  {theta_normal[1][0]:.4f}")
print(f"MktMilesFlown: {theta_normal[2][0]:.4f}")
print(f"Market_HHI:    {theta_normal[3][0]:.4f}")
print(f"Market_share:   {theta_normal[4][0]:.4f}")
print(f"MktCoupons:    {theta_normal[5][0]:.4f}")
print("training time: ", training_time)
```

```python
# Predict on training data
def predict(X, theta):
  X_b = np.c_[np.ones((X.shape[0], 1)), X]
  return X_b.dot(theta)

start_pred = time.time()
y_pred_normal = predict(X, theta_normal)
end_pred = time.time()
prediction_time_total = end_pred - start_pred

rows = X.shape[0]
pred_time_per_1k = prediction_time_total / rows * 1000
print("prediction time per 1k rows: ", pred_time_per_1k)
```

```python
from sklearn.metrics import mean_squared_error

mse_normal = mean_squared_error(y, y_pred_normal)
mae = mean_absolute_error(y, y_pred_normal)
r2 = r2_score(y, y_pred_normal)
print(f"Mean Squared Error on training data = {mse_normal:.4f}")
print(mae)
print(r2)
```

GRADIENT DESCENT

```python
def linear_regression_gradient_descent(X, y, learning_rate, num_iterations):
  # Implements Batch Gradient Descent for Linear Regression.
  # Takes, as arguments, the feature matrix (X), the target values (y), the learning rate, and teh number of gradient descent iterations
  # Outputs the trained coefficients and the MSE over iterations (cost_history)

  m = X.shape[0]

  # Add intercept term
  X_b = np.c_[np.ones((m, 1)), X]

  # Initialize weights randomly
  theta = np.random.randn(X_b.shape[1], 1)

  cost_history = []

  for i in range(num_iterations):
    gradients = (2/m) * X_b.T.dot(X_b.dot(theta) - y)
    theta -= learning_rate * gradients

    predictions = X_b.dot(theta)
    cost = np.mean((predictions - y) ** 2)
    cost_history.append(cost)

    if i % 100 == 0 or i == num_iterations - 1:
      print(f"Iteration {i:4d}: MSE = {cost:.4f}")

  return theta, cost_history
```

```python
from sklearn.preprocessing import StandardScaler

# Feature scaling
scaler_X = StandardScaler()
X_scaled = scaler_X.fit_transform(X)

# Target scaling is optional, usually not done unless y has large range
y = y.reshape(-1, 1)  # ensure it's 2D for matrix ops

theta_gd, cost_history = linear_regression_gradient_descent(X_scaled, y, 0.1, 1000)
```

```python
plt.figure(figsize=(10, 5))
plt.plot(range(len(cost_history)), cost_history, color='darkred')
plt.title("Cost Function vs. Iterations", fontweight='bold')
plt.xlabel("Iteration")
plt.ylabel("Mean Squared Error (MSE)")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
print("Coefficients from Gradient Descent (compared with normal equation results):")
print(f"Intercept:     {theta_gd[0][0]:.4f}, Difference = {abs(theta_normal[0][0]-theta_gd[0][0]):.4f}")
print(f"NonStopMiles:  {theta_gd[1][0]:.4f}, Difference = {abs(theta_normal[1][0]-theta_gd[1][0]):.4f}")
print(f"MktMilesFlown: {theta_gd[2][0]:.4f}, Difference = {abs(theta_normal[2][0]-theta_gd[2][0]):.4f}")
print(f"Market_HHI:    {theta_gd[3][0]:.4f}, Difference = {abs(theta_normal[3][0]-theta_gd[3][0]):.4f}")
print(f"Market_share:  {theta_gd[4][0]:.4f}, Difference = {abs(theta_normal[4][0]-theta_gd[4][0]):.4f}")
print(f"MktCoupons:    {theta_gd[5][0]:.4f}, Difference = {abs(theta_normal[5][0]-theta_gd[5][0]):.4f}")
```

```python
y_pred_gd = predict(X_scaled, theta_gd)

# 5. Evaluate
from sklearn.metrics import mean_squared_error
mse_gd = mean_squared_error(y, y_pred_gd)
print(f"Mean Squared Error from Gradient Descent = {mse_gd:.4f}")
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

# Step 1: Prepare your full dataset
X_full = df[['NonStopMiles','MktMilesFlown','Market_HHI', 'Market_share', 'MktCoupons']].values
y_full = df["Average_Fare"].values.reshape(-1, 1)

# Optional: log transform if you're using that model
# y_full = np.log1p(y_full)

# Step 2: Hold out a fixed test/validation set
X_train_full, X_val, y_train_full, y_val = train_test_split(X_full, y_full, test_size=0.2, random_state=42)

# Step 3: Scale the features (fit on full train set)
scaler = StandardScaler()
X_train_full_scaled = scaler.fit_transform(X_train_full)
X_val_scaled = scaler.transform(X_val)

# Step 4: Train on increasing data sizes
train_sizes = np.linspace(50000, 700000, 10, dtype=int)  # e.g. 100000, 200000, ...
train_errors = []
val_errors = []

for m in train_sizes:
    # Subset the training set
    X_train_m = X_train_full_scaled[:m]
    y_train_m = y_train_full[:m]

    # Train model using gradient descent
    theta_m, _ = linear_regression_gradient_descent(X_train_m, y_train_m, learning_rate=0.01, num_iterations=500)

    # Predict on training subset
    y_train_pred = predict(X_train_m, theta_m)
    train_mse = mean_squared_error(y_train_m, y_train_pred)

    # Predict on validation set
    y_val_pred = predict(X_val_scaled, theta_m)
    val_mse = mean_squared_error(y_val, y_val_pred)

    # Store errors
    train_errors.append(train_mse)
    val_errors.append(val_mse)

    print(f"Size {m}: Train MSE = {train_mse:.2f}, Val MSE = {val_mse:.2f}")
```

```python
plt.figure(figsize=(10, 6))
plt.plot(train_sizes, np.sqrt(train_errors), 'o-', label="Training RMSE", color='blue')
plt.plot(train_sizes, np.sqrt(val_errors), 'o-', label="Validation RMSE", color='green')

plt.xlabel("Training Set Size", fontsize=12)
plt.ylabel("RMSE", fontsize=12)
plt.title("Learning Curve (Linear Regression Gradient Descent)", fontsize=14)
plt.legend()
plt.grid(True)
plt.xlim(0, 800000)
plt.ylim(46.5, 48)

# Use full numbers on x-axis
# plt.xticks(train_sizes, labels=[f"{x:,}" for x in train_sizes])  # Formats like 100,000

plt.tight_layout()
plt.show()
```
